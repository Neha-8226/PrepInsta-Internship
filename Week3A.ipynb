Filling the null values in the age column

```python
new_df.Age.fillna(new_df.Age.mean(), inplace = True)
new_df['Age'] = new_df.Age.round(decimals = 1)
new_df
```

```python
new_df
```

```python
new_df.Salary
```

Formatting the Salary Column

```python
new_df['Salary'] = new_df.Salary.str.replace("$", " ")
new_df['Salary'] = new_df.Salary
print(new_df)
```

```python
new_df['Salary'] = new_df.Salary.str.replace("k", "000")
new_df['Salary'] = new_df.Salary
print(new_df)
```

Standardizing the Location column.

```python
new_df['Location'] = new_df.Location.str.replace(",In", " ")
new_df['Location'] = new_df.Location
print(new_df)
```

```python
new_df['Location'] = new_df.Location.str.replace(",Ny", " ")
new_df['Location'] = new_df.Location
print(new_df)
```

```python
new_df['Location'] = new_df.Location.str.replace(",Aus", " ")
new_df['Location'] = new_df.Location
print(new_df)
```

```python
new_df['Location'] = new_df.Location.str.replace(" In", " ")
new_df['Location'] = new_df.Location
print(new_df)
```

```python
new_df['Location'] = new_df.Location.str.replace(" Aus", " ")
new_df['Location'] = new_df.Location
print(new_df)
```

Formatting the rating column

```python
new_df.Rating.fillna(new_df.Rating.mean(), inplace = True)
new_df['Rating'] = new_df.Rating.round(decimals = 1)
new_df
```

```python
new_df['Rating'] = new_df['Rating'].abs()
new_df
```

```python
new_df['Rating'] = new_df.Rating.replace("1.0", "0.0")
new_df['Rating'] = new_df.Rating
print(new_df)
```

Analyzing the Easy Apply column

```python
new_df.rename(columns={"Easy Apply": "Apply"}, inplace=True)
new_df
```

```python
new_df['Apply'] = new_df.Apply.replace("-1", "FALSE")
new_df['Apply'] = new_df.Apply
print(new_df)
```

Exploring the Established Column

```python
new_df['Established'] = new_df['Established'].replace(-1, np.nan)
new_df
```

```python
new_df['Established'].mode
```

```python
new_df['Established'] = new_df['Established'].replace(np.nan, new_df['Established'].mode().iloc[0])
new_df
```

```python
new_df['Established'] = new_df['Established'].astype(int)
new_df
```
