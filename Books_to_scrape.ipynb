Importing the libraries

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

Connecting the website to the Colab notebook

```python
url = 'https://books.toscrape.com/'
response = requests.get(url)
```

```python
print(response.text)
```

```python
soup = bs(response.text)
type(soup)
```

```python
soup.find('title').text.strip()
```

```python
books_tag = soup.find_all('article',class_='product_pod')
```

```python
len(books_tag)
```

Extracting the HTML code for one book

```python
book = books_tag[0]
print(book)
```

Extracting the title for the books

```python
title_tag = book.find('a',title=True)
title_tag
```

```python
title_tag.text
```

```python
title_tag['href']
```

```python
title_tag['title']
```

Extracting all the other titles

```python
all_titles = []
for book in soup.find_all("article", class_="product_pod"):
    title_tag = book.find("a", title=True)
    if title_tag:
        title = title_tag.text.strip()
        title = title_tag['title']
        all_titles.append(title)
```

```python
print(all_titles)
```

Extracting the Ratings of the books

```python
rating = book.find('p')['class']
```

```python
rating[1]
```

Extracting all the ratings

```python
all_ratings = []
for book in soup.find_all("article", class_="product_pod"):
    rating_element = book.find("p", class_="star-rating")
    if rating_element:
        rating = rating_element["class"][1]
        all_ratings.append(rating)
```

```python
print(all_ratings)
```

Extracting the prices of the books

```python
price = book.find("p", class_ = "price_color")
```

```python
price
```

```python
all_prices = []
for book in soup.find_all("article", class_="product_pod"):
    price_element = book.find("p", class_="price_color")
    if price_element:
        price = price_element.text.strip()
        all_prices.append(price)
```

```python
print(all_prices)
```

Scrapping the links

```python
link = book.find("article", class_ = "product_pod")
```

```python
all_links = []
for book in soup.find_all("article", class_="product_pod"):
    link_tag = book.find("a")
    if link_tag:
        link = url + link_tag["href"]
        all_links.append(link)
```

```python
print(all_links)
```

Combining all the lists together

```python
book_data = {
    "title": all_titles,
    "rating": all_ratings,
    "price": all_prices,
    "link": all_links
}          #Creating a dictionary
```

```python
df = pd.DataFrame(book_data)    # Creating a dataframe
```

```python
df.columns = ["Title", "Rating", "Price", "Link"]  #Naming the columns
```

```python
print(df.to_string())
```

```python
df.to_csv("books_data.csv", index=False)   # Saving as a csv file
```
