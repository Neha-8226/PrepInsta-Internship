Checking for null values

```python
super.info()
```

```python
super['Postal Code'].isna().sum()
```

Here we find that the Postal Code column has 11 null values

So we make a list new variable for storing all the null values

```python
null_mask = super['Postal Code'].isnull()
```

Once the variable is created we make a list out of it and get the indexes of the null values

```python
null_indexes = null_mask[null_mask == True].index.tolist()
```

```python
print(super.loc[null_indexes, "Postal Code"])
```

After getting the indexes we get the row number's information from the dataset and find that all the null values belong to the city Burlington, Vermont and the postal code for the same is '05401'.

```python
row_2234 = super.iloc[2234]
print(row_2234)
```

We then replace the missing values with the Postal Code '05401' and get rid  of null values.

```python
super['Postal Code'] = super['Postal Code'].fillna(value = '05401')
```

```python
super['Postal Code'].isna().sum()
```

```python
super['Postal Code'].head()
```

Now we change the data type of the column from float to integer as Postal Code cannot be a float value

```python
super['Postal Code'] = super['Postal Code'].astype(int)
```

```python
super['Postal Code'].dtype
```

Creating visuals

1. Sales Distribution

```python
sns.histplot(super["Sales"], log_scale = True, palette = 'coolwarm')   # Creating a histogram for depicting the sales.
plt.title("Sales Distribution")
```

2. Relationship between Order Date and Sales

```python
sns.set_palette("coolwarm")
```

```python
sns.barplot(x="Order Date", y="Sales", data=super, palette = 'muted')  #Creating a scatter plot for showing the relationship between Order Date and Sales
plt.xlabel("Order Date")
plt.ylabel("Sales")
plt.title("Sales Trend over Time")
```

3. Relation between Order Date and Sales by Customer Segment

```python
plt.figure(figsize=(10, 4))
sns.scatterplot(x='Order Date', y='Sales', hue='Segment', data=super, s=100, palette = 'muted')
plt.title('Sales Trend overtime by Customer Segment')
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.legend(title='Customer Segment')
plt.show()
```

The Consumer Segment is slightly separable and higher than the other two sectors in comparison to sales by Customer Segment.

4. Correlatinal Heatmap

```python
sns.heatmap(super.corr(), annot = True, cmap = 'coolwarm', linewidth = .5)
plt.title("Correlation Heatmap")
```

5. Count of Orders by Ship Mode

```python
sns.set_palette("coolwarm")
```

```python
sns.countplot(x = "Ship Mode", data = super)
plt.title("Count of Orders by Ship Mode")
```

Most orders were shipped by Standard Class. The number of orders shipped on the same day was comparatively the lowest.

6. Sales Distribution across Product Categories

```python
plt.figure(figsize=(10, 5))
plt.boxplot([super['Sales'][super['Category'] == category] for category in super['Category'].unique()],
            labels=super['Category'].unique())
plt.title('Sales Distribution by Product Category')
plt.xlabel('Product Category')
plt.yscale('log')
plt.ylabel('Sales(Log Scale)')
plt.show()
```

7. Sales Trends for Different Customer Segments

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x="Order Date", y="Sales", hue="Segment", data=super, palette = 'hls')
plt.title("Sales Trends by Customer Segment")
```

The sales for the Consumer Segment is huge than the Home Office Segment which is slightly higher than the Corporate Segment.

8. Sales patterns across Regions and Countries

```python
sns.barplot(x="Region", y="Sales", hue="Country", data=super, palette = 'hls')
plt.title("Sales by Region and Country")
```

The Sales in the South and East region are nearly equal followed by West region and Central Region.

9. Plot for Sales vs PostalCode

```python
plt.figure(figsize=(12, 7))
pairplot = sns.pairplot(super[['Sales', 'Postal Code']], height=4, aspect=1.5, palette = 'muted')
pairplot.fig.suptitle('Sales by Postal Code', y=1.02)
plt.show()
```

10. Count of Products by Category

```python
sns.countplot(x='Category', data=super, palette = 'hls')
plt.title('Count of Products in Each Category')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.show()
```

This shows that the Count of Products of Office Supplies is the highest followed by Furniture and Technology.
